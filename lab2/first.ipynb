```python
def is_valid_word(s):
    # Define sets
    lowercase_letters = set('abcdefghijklmnopqrstuvwxyz')

    # States
    q0 = 'start'
    q1 = 'valid_word'
    qr = 'reject'

    # Initial state
    state = q0

    for ch in s:
        if state == q0:
            if ch in lowercase_letters:
                state = q1
            else:
                state = qr
        elif state == q1:
            if ch in lowercase_letters:
                state = q1
            else:
                state = qr
        elif state == qr:
            # Stay in reject once entered
            state = qr

    # Accept if in q1 at the end
    if state == q1:
        return "Accepted"
    else:
        return "Not Accepted"

# Test cases
test_strings = ["cat", "dog", "a", "zebra", "dog1", "1dog", "DogHouse", "Dog_house", " cats"]

for word in test_strings:
    print(f"{word}: {is_valid_word(word)}")
```

```python

lowercase_letters = set('abcdefghijklmnopqrstuvwxyz')

transitions = {
    "q0": {letter: "qf" for letter in lowercase_letters},
    "qf": {letter: "qf" for letter in lowercase_letters},  # stay in qf if more letters come
    "qr": {letter: "qr" for letter in lowercase_letters},  # stay in qr if entered
}

dfa = VisualDFA(
    states={"q0","qf","qr"},
    input_symbols=lowercase_letters,
    transitions=transitions,
    initial_state="q0",
    final_states={"qf"},
)

dfa.show_diagram(path='dfa_diagram.png')
```
