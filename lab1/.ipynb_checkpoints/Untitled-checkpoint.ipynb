```python
import re 
```

# load the dataset

```python
with open ("gu.txt", encoding="utf-8") as file:
    lines=[]
    for i in range(500):
        line = file.readline()
        lines.append(line)
    
    # lines = file.readlines()
```

# sentence tokenizer

```python
def sentence_tokenizer(line):
    pattern = r'.+?[.?!]|.+$'
    return re.findall(pattern,line)
    
```

```python
sentences = []

for i,lin in enumerate(lines):
    if i%2==0:
        
        sentence = sentence_tokenizer(lin)
        sentences.extend(sentence)  

# for s in sentences:
#     print(s)

print(sentences) 

```

```python
def word_tokenizer(text):
    pattern = re.compile(r'''
        # URLs
        https?://[^\s]+ |
        www\.[^\s]+ |

        # Email IDs
        [\w.+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+ |

        # Dates (formats: DD/MM/YYYY, DD-MM-YYYY, DD.MM.YYYY, etc.)
        \d{1,2}[/-]\d{1,2}[/-]\d{2,4} |
        \d{1,2}[ ]?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[ ]?\d{2,4} |

        # Gujarati numbers with decimals (૧૨.૩૪)
        [\u0AE6-\u0AEF]+(?:\.\d+)? |

        # English numbers with decimals (12.34)
        \d+(?:\.\d+)? |

        # Gujarati words (letters and signs)
        [\u0A80-\u0AFF]+ |

        # Punctuation
        [.,!?;:'"()\[\]{}<>।|—-]
    ''', re.VERBOSE | re.IGNORECASE)
    
    return pattern.findall(text)
```

# word tokenizer

```python
words= []

for s in sentences:
    words.extend(word_tokenizer(s))


```

# writing into the file

```python
with open("temp-guj-sent.txt", "w", encoding="utf-8") as f:
    for line in sentences:
        f.write(line + "\n")
```

# corpus statistics

```python
def corpus_statistics(sentences,words):
    total_sentences = len(sentences) 
    total_words = len(words) 
    total_characters = sum(len(token) for token in words if token.isalnum()) 
    avg_sentence_length = total_words/total_sentences if total_sentences>0 else 0 
    avg_word_length = total_characters/total_words if total_words>0 else 0 

    unique_tokens = set(words) 
    ttr = len(unique_tokens)/total_words if total_words>0 else 0 

    stat_dict = {
       "Total number of sentences" : total_sentences,
       "Total number of words" : total_words, 
       "Total number of characters" : total_characters, 
       "Average Sentence Length" : avg_sentence_length, 
       "Average word length" : avg_word_length, 
       "TTR" : ttr 
                }
    return stat_dict
```

```python
stat  = corpus_statistics(sentences,words) 
```

```python
for key,value in stat.items(): 
    print(key," : ",value) 
```
